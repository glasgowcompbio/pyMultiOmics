```python
name_id_dict = {}
for key in gene_name_dict.keys():
    if key[0:4] == 'ENSG':
        name_id_dict[gene_name_dict[key]] = key

```

```python
len(name_id_dict)
```

```python
name_id_df = pd.DataFrame({"name":name_id_dict.keys(),"Identifier":name_id_dict.values()})
name_id_df
```

```python
df = pd.read_csv("/Users/wangkeqing/Desktop/master-gla/project/data/covid_data/covid_genes_include_p.csv", sep=",")
df
```

```python
joined_df = df.join(name_id_df.set_index('name'), on='name')
joined_df
```

```python
col_list=[]
for col in joined_df.columns:
    col_list.append(col)
```

```python
col_list[0]='Identifier'
del col_list[-1]
```

```python
joined_df = joined_df[col_list]
joined_df
```

```python
new_col_name = []
for col in joined_df.columns:
    if str(col)=='id':
        new_col_name.append('Identifier')
    else:
        new_col_name.append('sample_'+str(col))
joined_df.columns = new_col_name
joined_df
```

```python
drop_row_list = []
for index, row in joined_df.iterrows():
    if str(row['Identifier']) == 'nan':
        drop_row_list.append(index)
```

```python
len(drop_row_list)
```

```python
joined_df.drop(drop_row_list, axis=0, inplace=True)
joined_df
```

```python
joined_df.to_csv("/Users/wangkeqing/Desktop/master-gla/project/data/covid_data/covid_genes_include_p.csv", index=False, header=True,sep=",")
```
